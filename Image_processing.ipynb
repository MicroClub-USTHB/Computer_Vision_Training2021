```
import matplotlib.pyplot as plt
import numpy as np 
import cv2 
from skimage import io
```

```
#load image
image= io.imread('https://babzman.com/wp-content/uploads/2014/03/maqam.jpg')
#print image as is ...matrix
print(image)

```

```
print("shape of image {}  ".format(image.shape))
```

```
#plot image
plt.imshow(image)
```

```
image1=cv2.flip(image,0)

fig= plt.figure(figsize=(10,10))
fig.add_subplot(121)
plt.imshow(image)
fig.add_subplot(122)
plt.imshow(image1)
```

```
#load image

image2= io.imread('https://babzman.com/wp-content/uploads/2014/03/maqam.jpg',True)
plt.figure(figsize=(10,10))
plt.imshow(image2,cmap='gray')

```

```
filter = np.ones((5,5),np.float64)/25 #define filter
```

```
f1=np.ones((5,5),np.float64)
print(f1)
f2=f1/25
print(f2)
```

```
filtered_image= cv2.filter2D(image2,-1,filter) #cv2 function to apply filters
figure=plt.figure(figsize=(10,10))
plt.imshow(filtered_image,cmap='binary')
```

```
filtered_image1=cv2.GaussianBlur(image,(9,9),0) #cv2 function for blur
plt.figure(figsize=(10,10))
plt.imshow(filtered_image1)
```

```
test_image=cv2.imread('Yann.jpg',0)
filtered_image2= cv2.Canny(test_image,100,200) #cv2 function for edge detection
plt.figure(figsize=(10,10))
plt.imshow(filtered_image2,cmap='gray')
```
