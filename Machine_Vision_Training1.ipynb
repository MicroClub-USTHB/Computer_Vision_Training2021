 # **2-Computer Vision: the core concepts**

## What is Computer Vision ?  

Computer vision is artificial intelligence applied to the visual world. It enables us to replicate the abilities of the human eye and the visual cortex.
The core challenge thus is to give computers the ability to see and more importantly to understand what they see 

> **" Just like to hear is not the same as to listen, to take pictures is not the sale as to see and by to see we really mean to make understanding "** - Fei Fei Li 

To make understainding out of the pictures taken means :
* know what's in the image 
* know where these things are in the image 
* know the structure/ composition of the image 
* know the relation between the things in the image 
* know how to explain/describe the scenario represented in the image.

## What is Image processing ?

> **Image processing:**taking an image as is and alterating it content thus changing is end result/ appearance. [ example: reshape, rescale, threshholding..] 
Editing is the most obvivous aspect of image processing ... it is done through filters.

> **Filters:** matrices of chosen specific numbers that produce a wanted effect on the image at hand .... every image editing filter has its own set matrice (blurry, smoothing,

> **Confusion:** some effects of image processing are mistaken for computer vision such as edge detection. but once again we need to understand that the core idea of computer vision is for the *machine* to be aware of the content of the image, NOT the viewer.

> **Fusion:** Although they are set to be two different and independent domains, image processing and computer vision do work hand in hand sometimes which gave birth to smart filters (social media effects) but more importantly because in some case data need to be processed before passing it through the deep learning model and inside the network.

## How are images represented/stored ?

We said that filters are matrices but what about images ? ... well same ! 

* Images are matrices in which each element is called a pixel (smallest non dividable element of the picture). 
* The values of the pixels inside the matrix range from 0 to 255.
* The value of 0 to 255 represents the scal of grey in a GreyScale representation. 
* The value of 0 to 255 represents the scale of Red,Green,Blue in a RGB representation
* B&W images are formed of one unique channel (matrix)
* Colored images are formed of 3 channels (tensor)

## A brief history of neural networks

Well basically since computers were invented they have been super good at computing ... pretty logical given their name! Yet these super calculators were still enable to solve the most basic human tasks such as the famous cat vs dog classification. 

This pushed engineers to rethink their approach : *what if instead of trying to build a machine we tried to build a brain ?*

They studied the brain, the communication of a signal between neurons following a specific path through the nervous system. this then gave birth to what is known as a black box of computation : neural networks. 

In what follows we'll take a look at what happens inside the black box. Spoiler alert there's math everywhere.


![](https://hackernoon.com/hn-images/1*pPYFYOZ-8dja-kRrkmuGOQ.jpeg)

#### **The Fully Connected Network architecture**
The Fully connected network (or multilayer perceptron) as shown in this picture is a forward connection of nods distributred throughout a number of layers. these layers are of 3 categories: 

* the input layer: introduces the data into the network
* the output layer: communicates the result of the calculations 
* the hidden layers: a number of layers between the input and output that forms the core of the computation power of the sturcture, deferentiating shallow networks from deep ones. 


![](https://www.kdnuggets.com/wp-content/uploads/neural-network-input-hidden-output.jpg)

#### **Computation at the neuron level**

To understand the functioning of a network it is important to understand how each neuron works. In short we can summurize it as a summation of weighted values passed through an activation function.


$ y=  f( \sum ( w . x) +b) $ .  Where
-  x is the input vector of the neuron
-  w is the vector of weights associated to each arrow
-  b is the bias 
-  f is the activation function 
-  y is the output of the neuron



![](https://ichi.pro/assets/images/max/724/1*xYCVODGB7RwJ9RynebB2qw.gif) 

So as we can see, the mission of a single neuron is to 1- weight 2- sum up and finally 3- activate. 
the 2 first steps are common to multiple machine learning algorithms such as linear regression. 


* **The activation function** is what sets the neural network algorithme apart. The goal of this function is to bring non linearity into the calculus made by the neuron thus allowing the network to solve non-linearly separable classifications by deciding if an information will be forwarded to the next layer or not. From a biologic point of view it is comparable to the potential of activation that decideds if a respons is fired after a signal is received.   







### **Communication throughout the network**

Out of the list of parameters included in the privious equation we can distinguish two categories: learned parameters and hyper parameters. the later describes a set of decisions made by the programmer beforehands depending on the purpose of the model developed. these decision include the choice of the activation function (sigmoid, softmax, relu..) [ and also picking the learning rate, loss function,number of layers/neurons, optimizer ]

As for the learnd parameters: weights and bias of each neuron... 

#### Learning process:

The overall instinct of a supervised algorithm is to provide a hypotethis given an input data then compare that hypotethis to the expected output (aka the label). In a neural network this process is done through two steps *feedforward* and *backpropagation*. 

* feedforward: the forward pass of information through the different layers of the network from input to output creating a cascade of calculus that produces a final value: the estimated result by the network (hypotethis). Once the hypotethis is available, it is compared to the expected result through the calculation of the loss function. 
* backpropagation: the backword pass over each neuron of the network starting from the last one to update the weights according to the loss obtained after the feedforward. The weights are updated using the gradient decent algorithm.


$ W '=W - \alpha (\frac {\partial Error}{\partial W} )$

![](https://community.alteryx.com/t5/image/serverpage/image-id/42339i8BA3F2CCCEDE7458?v=1.0)


The learning process is repeated a number of times (epochs) untill the loss obtains a relatively low value. With the final weights a good model will be sensitive to the data it was trained to process, meaning each neuron is now able to recognize a pattern it has learned. in the case of a classification, one data will trigger a specific path where the class most likely to be true will be excited thus providing the right classification.

### Quick example of tf code
Here we will take a look at a number of tf code lines that we'll be playing a lot with in this training camp. In this rather simple/short example we will be using a shallow fully connected network to convert Celsius into Fahrenheit.
The math formula that allows to do this conversion is as follows **F= C x 9/5 + 32** . Of course our network will not be given this formula. The network will be given a set of data that gathers temperatures in celsius and in fahrenheit to follow a supervised training.

```
# 1- import the dependencies
import tensorflow as tf 

# 2- get the data
celsius=[0,-10,45,20,-5,-3,25,12,30,29,22]
fahrenheit=[32,14,113,68,23,26.6,77,53.6,86,84.2,71.6]

 # 3- build the network
model0=tf.keras.Sequential([
                            tf.keras.layers.Dense(4,input_shape=[1]),
                            tf.keras.layers.Dense(4),
                            tf.keras.layers.Dense(1)
])

# 4- pick the hyperparameters
model0.compile(loss=tf.keras.losses.mean_squared_error,optimizer=tf.keras.optimizers.Adam(0.1))

# 5- train the model
history=model0.fit(x=celsius,y=fahrenheit,epochs=800,verbose='false')
```

```
# 6- test the model 
print(model0.predict([36]))
```

### Second example of fully connected networks with tensorflow
Now we'll take another look at some tf code on what is called the "Hello World" of deep learning... Fashion Mnist clothing classification ! 


**Fashion Mnist** is an open dataset gathering 70000 images of clothing spread over 10 classes : T-shirt/top, Trouser, Pullover, Dress, Coat,
               Sandal, Shirt, Sneaker, Bag, Ankle boot.

In what follows we'll train a small FC network to classify these pieces of clothing based on the 28x28 images of the dataset.

```

#1- import dependencies 
import numpy as np
import matplotlib.pyplot as plt

#2- get the data
MNIST=tf.keras.datasets.fashion_mnist
(x_train,y_train),(x_test,y_test)=MNIST.load_data()
x_train=x_train/255.0
x_test= x_test/255.0

#3- build the network
model1=tf.keras.Sequential([
                            tf.keras.layers.Flatten(input_shape=(28,28)),
                            tf.keras.layers.Dense(128,activation="relu"),
                            tf.keras.layers.Dense(128,activation="relu"),
                            tf.keras.layers.Dense(10)
])
#4- compile the model
model1.compile(optimizer='Adam',loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics='accuracy')
#5- train the model 
history=model1.fit(x=x_train,y=y_train,epochs=10)

```

```
#6- test the model
loss,acc=model1.evaluate(x_test,y_test)
 
```

```
#let's take a closer look at what this model1 actually does : 
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
plt.imshow(x_test[26],cmap=plt.cm.binary)
predict=model1.predict(x_test)
prediction=np.argmax(predict[26])
prediction_class=class_names[prediction]
label_id= y_test[26]
label=class_names[label_id]
plt.title(prediction_class,color='blue',loc='left')
plt.title(label,loc='right')
plt.show()
```
