### **L'architecture Fully Connected**
L'architecture Fully Connected (ou multilayer perceptron ou feedforward network) se présente comme une succession de couches entièrement connectées entre elles. Les couches quant à elles sont des regroupements de neurones qui sont l'unité de calcule du réseau (où la magie opère).
Les couches du réseau FC sont de 3 types selon leur position: 

* Couche d'entrée: représente le point d'accès des données vers le réseau 
* Couche de sortie: transmet le résultats finale produit par le réseau 
* Couches cachées: un nombre de couches situées entre l'input et l'output. (D'où le nom "caché" parce qu'elles ne sont pas en contacte directe avec le "monde exterieur"). Les couches caché n'ont donc aucun role de "communication" mais uniquement de calcule. C'est à ce niveau là que la compléxité du réseau est "décidé",c'est d'ailleur le grand nombre de couches cachées qui caractérise les réseaux de neurones profonds.

### **Les calculs effectué par un neurone**

Pour comprendre le fonctionement du réseau de neurone il est important de comprendre ce qui se passe qu niveau d'un seul neurone qui est comme on l'a dit plus haut "l'unité de calcule". Pour résumer on peut dire que la sortie d'un neurone est une *somme* de valeurs *pondérées* que l'on passe par une *fonction d'activation* comme le montre la formule suivante : 


$ y=  f( \sum ( w . x) +b) $ .  tel que
-  x : le vecteur de valeurs d'entrée 
-  w : le vecteur de poids associés à chaque valeur
-  b : le biais
-  f : fonction d'activation
-  y : sortie du neurone



![](https://ichi.pro/assets/images/max/724/1*xYCVODGB7RwJ9RynebB2qw.gif) 

On peut donc diviser les calcule effectués par un neurone en 3 parties,  1- pondérer les valeurs reçu 2- faire la somme and enfin 3- **l'activation** . 
Les 2 premières étapes sont communes à plusieurs algorithmes de Machine Learning c'est l'activation du neurones qui fait l'innovation du réseau 




#### **Fonction d'activation**

* La fonction d'activation est l'équivalent du potentiel d'action.**Le potentiel d'action d'un neurone biologique** est ce qui lui permet de **décider si un signale transmit sera propagé ou pas**. Analogiquement la faction d'activation permetra de décider si une information (valeur numérique) sera partagé ou ignoré (exemple ReLU)


* Plus que ça l'objectif de cette fonction est aussi de "casser" la linéarité des calcules effectués et donc de permettre à un neurone de résoudre des problèmes qui ne sont pas forcément linéairement séparables (show shapes)

### **Communication à travers le réseau**

Parmi les différents paramèters inclus dans la formule decrivant les calcules d'un neurone, on distingue deux catégories:les paramètres à apprendre et les Hyper paramètres.
* Les hyper paramètres sont les paramètres choisis par le programeur avant meme le lancement de l'entrainement et qui vont bien évidemment influer sur les résultats et les performances de notre modèle. Parmi ces hyper parametre on aura la fonction d'activation (sigmoid, softmax, relu..) [ mais aussi le choix de la fonction perte, le choix du learning rate...]

* Les paramètres à apprendre quant à eux ne sont pas constant bien au contraire ils sont sujet au changement tout au long de l'apprentissage jusqu'à obtenir les "meilleures valeurs"

#### Processus d'apprentissage:

Comme on l'a mentionné plus tot le but du processus d'apprentissage sera de donner en sortie une hypothèse selon l'entrée transmise puis juger de l'exactitude de la réponse donnée selon le label qu'on a déja.
 On peut diviser ce processus d'apprentissage en deux étapes: *feedforward* et *backpropagation*. 

* feedforward: désignant le cheminement en cascade des données à travers les différentes couches de l'input vers l'output de manière consécutive. Cela signifie que la sortie d’une couche i sera l’entrée de la couche i+1. L'enchainement des différents calcules transmet en sortie un résulat final que l'on appel *hypothèse*. L'hypothèse sera par la suite comparé au résulat escompté (label) à l'aide de la fonction perte. 

* backpropagation: c'est le passage de l'information dans le sens inverse (output-> input) afin de corriger les poids appris dans chaque couche, proportionnellement à l’erreur enregistrée par le réseau suite au
feedforward. Les poids sont mis à jour à l'aide de l'algorithme de la déscente du gradient dans le but de minimiser l'erreur enregistrée.


$ W '=W - \alpha (\frac {\partial Error}{\partial W} )$

![](https://community.alteryx.com/t5/image/serverpage/image-id/42339i8BA3F2CCCEDE7458?v=1.0)


- Le processus d'apprentissage est donc une combinaison de ces deux étapes par lesquelles le modèle va passer un nombre de fois (epochs)jusqu'à ce que l'erreur soit assez petite.

-  Avec les meilleurs poids un bon model aura la prossibilité de prédire une sortie s'approchant du résultat attendu pour les données vu en entrainement mais surtout les données jamais vu (capacité de généralisation).

### Réseau de neurone avec Tensorflow
Dans les cellule qui suivent nous allons voir comment importer et utiliser Tensorflow pour créer et entrainer un réseau de neurone (at a high level).
Dans cet exemple on va mettre au point un réseau FC pour convertir les Celsius en Fahrenheit.
La formule permettant d'effectuer la conversion est la suivante **F= C x 9/5 + 32** . Bien évidement le modèle n'aura pas accès à cette formule il devra donc trouver son propre système de calcule en apprenant les bons paramètres. Pour y arriver le réseau de neurone aura comme Dataset un ensemble de températures en Celsius suivi de leurs équivalents en Fahrenheit.

```
# 1- import the dependencies
import tensorflow as tf 

# 2- get the data
celsius=[0,-10,45,20,-5,-3,25,12,30,29,22,-12,24]
fahrenheit=[32,14,113,68,23,26.6,77,53.6,86,84.2,71.6,10.4,75.2]

 # 3- build the network
model0=tf.keras.Sequential([  
                            tf.keras.layers.Dense(1,input_shape=[1]),
                            tf.keras.layers.Dense(4,activation='relu'),
                            tf.keras.layers.Dense(1)
])

# 4- pick the hyperparameters
model0.compile(loss=tf.keras.losses.mean_squared_error,optimizer=tf.keras.optimizers.Adam(0.01))

```

```
model0.summary()
```

```
print(model0.get_weights())
```

```
# 5- train the model
history=model0.fit(x=celsius,y=fahrenheit,epochs=800,verbose='false')
```

```
print(model0.get_weights())
```

```
# 6- test the model 
print(model0.predict([36]))
```
